```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
MSIZE = 5
NSIZE = 4
SIZE = MSIZE * NSIZE
```

```python
probs = np.fromfile('prob.bin', dtype=np.float32).reshape((-1,SIZE))
rs = np.fromfile('rsum.bin', dtype=np.float32).reshape((-1,SIZE))
result = np.fromfile('result.bin', dtype=np.float64).reshape((-1,NSIZE,MSIZE))
```

```python
titles = ["Rij", "Probabilities", "Mean Rij"]
fig, axx = plt.subplots(nrows=1, ncols=3, figsize=(15,5), sharey=True, sharex=True)
for i in range(3):
    axx[i].imshow(result[i], origin='lower', interpolation='none')
    axx[i].set_xticks(np.arange(MSIZE))
    axx[i].set_yticks(np.arange(NSIZE))
    axx[i].set_title(titles[i])
    axx[i].grid('on')
fig.tight_layout()
plt.show()
```

```python
fig, axx = plt.subplots(nrows=NSIZE, ncols=MSIZE, figsize=(3*MSIZE,3*NSIZE), sharey=True)
for i in range(SIZE):
    x = i % MSIZE
    y = i // MSIZE
    axx[y,x].plot(probs[::100,i])
    axx[y,x].set_ylim(0, np.max(probs))
    axx[y,x].set_xlim(0)
fig.tight_layout()
plt.show()
```

```python
fig, axx = plt.subplots(nrows=NSIZE, ncols=MSIZE, figsize=(3*MSIZE,3*NSIZE), sharey=True)
for i in range(SIZE):
    x = i % MSIZE
    y = i // MSIZE    
    axx[y,x].plot(rs[::100,i])
    axx[y,x].set_ylim(0, 1.1*np.max(rs))
    axx[y,x].set_xlim(0)
fig.tight_layout()
plt.show()
```
